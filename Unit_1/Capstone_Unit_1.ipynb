
## Agriculture Weed Control

Weed control is an integral component crop management in agriculture. Weeds compete for space, nutrients and sunlight and left unchecked the weeds will out compete and overwhelm the crops. Modern agriculture techniques employ the use of chemicals such as glyphosate to manage weeds. Recently these chemical methods have become scrutinized and even linked to cancer. Research into crop management utilizing multiple techniques such as the work done by the LONG-TERM ECOLOGICAL RESEARCH (LTER) PROGRAM gives insight into different management techniques. One such study at Michigan State University investigated different treatments for weed control. Detailed information and the complete data set can be found at  https://lter.kbs.msu.edu/datasets/7 <br>

The treatment study has 4 groups T1-T4, below is a brief overview of the study.
* Conventional (T1) and No-till (T2) treatments receive conventional levels of chemical inputs, but with chisel plow and no-till management, respectively. <br> 
* The Reduced Input (T3) and Biologically Based (T4) treatments are low-chemical/biologically-based input systems with a winter leguminous cover crop; T3 receives banded herbicide and starter N at planting and T4 (certified organic) receives no chemical inputs nor compost or manure. Both T3 and T4 receive additional post-planting cultivation and T4 is rotary-hoed to control weeds.  
 
Exploratory analysis of the data is shown below, it can be seen that the no-chemical treatment T4 lags behind compared to the chemically treated crops. We hypothesize that you can accomplish weed control comparable to the groups using a chemical method.
  
Here we propose an experiment comparing a new treatment group using a High density row planting scheme and "No-Till" methods, where weed control is accomplished using a thermal treatment. The treatment will use a burner to kill weeds before they become established. The Hypothesis is that the Thermal treatment method can be as effective as the chemical method for managing weeds.

The experimental plan is to replace treatment T4 with the new treatment. The data collected will be weekly image data similar to the Kaggle data set https://www.kaggle.com/fpeccia/weed-detection-in-soybean-crops <br>
this will allow for more sampling than yearly crop yields that can have bias from factors besides weeds that can influence the crop yields.

The analysis will determine how effective you can manage weed growth by comparing the image data of the two treatments. Weed density and diversity will be compared between the two treatments and effective treatment is considered to where weed growth is managed to emerging plants and not allowed to mature.


```python
# import all necessary libraries 
import plotly.plotly as py
import pandas as pd 
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 
  
# to get the connection 
init_notebook_mode(connected = True) 

# Loading an example dataset provided by the seaborn package.

df = pd.read_csv('51_agro.csv')

T1 = df.loc[(df['Treatment']=='T1')] 
T2 = df.loc[(df['Treatment']=='T2')] 
T3 = df.loc[(df['Treatment']=='T3')] 
T4 = df.loc[(df['Treatment']=='T4')] 

T1.head()
```

```python
# Create Scatter Plot
trace1 = go.Scatter(x = T1['Year'], y = T1['Yield_bu_A'], mode = 'markers', name = 'T1')
trace2 = go.Scatter(x = T2['Year'], y = T2['Yield_bu_A'], mode = 'markers', name = 'T2')
trace3 = go.Scatter(x = T3['Year'], y = T3['Yield_bu_A'], mode = 'markers', name = 'T3')
trace4 = go.Scatter(x = T4['Year'], y = T4['Yield_bu_A'], mode = 'markers', name = 'T4')

layout = go.Layout(title='Yield of Different Treatments per Year',
                      xaxis=dict(title ='Year'), 
                       yaxis=dict(title='Total Production')
                  )

data = [trace1, trace2, trace3, trace4]
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='Scatter-plots')
```

```python
# Create Box Plots of the Different Treatments
trace1 = go.Box(x = T1['Treatment'], y = T1['Yield_bu_A'], name = 'T1')
trace2 = go.Box(x = T2['Treatment'], y = T2['Yield_bu_A'], name = 'T2')
trace3 = go.Box(x = T3['Treatment'], y = T3['Yield_bu_A'], name = 'T3')
trace4 = go.Box(x = T4['Treatment'], y = T4['Yield_bu_A'], name = 'T4')

layout = go.Layout(title='Overall Yield of Different Treatments',
                      xaxis=dict(title ='Treatment'), 
                       yaxis=dict(title='Total Production')
                  )

data = [trace1, trace2, trace3, trace4]
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='Box-plots')
```
