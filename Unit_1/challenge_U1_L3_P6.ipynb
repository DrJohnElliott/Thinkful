```python
import pandas as pd
import re
import statistics as st
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding="mac_roman")
df.head()

# Capitalize Journal Names
df['Journal title'] = df['Journal title'].apply(lambda x: str(x).capitalize())

# Rename long varible name
df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)': 'price'}, inplace=True)

# scrub symbols from price data
df['price'] = df['price'].apply(lambda x: x.replace('£', ''))  
df['price'] = df['price'].apply(lambda x: x.replace('$', ''))

# remove rows that are not aligned
df['str_len'] = df['price'].apply(lambda x: len(x))
df = df[df.str_len <= 7]

# try to convert price data to float
df['price'] = df['price'].astype('float64')
#df['price'] = pd.to_numeric(df.price)

# Check data is converted
df['is_digit'] = df['price'].apply(lambda x: str(x).isnumeric())

# Journal Counts
counts = df[['Journal title','Article title']].groupby(['Journal title'])['Journal title'].count().nlargest(5)

#df.head()
my_data =  pd.DataFrame(counts)
df.head()
```

```python
# Next, calculate the mean, median, and standard deviation 
# of the open-access cost per article for each journal 
```

```python
# Mean
mymean1 = df['price'][df['Journal title'] == counts.index[0]].mean()
mymean2 = df['price'][df['Journal title'] == counts.index[1]].mean()
mymean3 = df['price'][df['Journal title'] == counts.index[2]].mean()
mymean4 = df['price'][df['Journal title'] == counts.index[3]].mean()
mymean5 = df['price'][df['Journal title'] == counts.index[4]].mean()

# Median
mymedian1 = st.median(df['price'][df['Journal title'] == counts.index[0]])
mymedian2 = st.median(df['price'][df['Journal title'] == counts.index[1]])
mymedian3 = st.median(df['price'][df['Journal title'] == counts.index[2]])
mymedian4 = st.median(df['price'][df['Journal title'] == counts.index[3]])
mymedian5 = st.median(df['price'][df['Journal title'] == counts.index[4]])


# STD
mystd1 = df['price'][df['Journal title'] == counts.index[0]].std()
mystd2 = df['price'][df['Journal title'] == counts.index[1]].std()
mystd3 = df['price'][df['Journal title'] == counts.index[2]].std()
mystd4 = df['price'][df['Journal title'] == counts.index[3]].std()
mystd5 = df['price'][df['Journal title'] == counts.index[4]].std()
```

```python
my_data['mean'] = [mymean1, mymean2, mymean3, mymean4, mymean5]
my_data['median'] = [mymedian1, mymedian2, mymedian3, mymedian4, mymedian5]
my_data['stdev'] = [mystd1, mystd2, mystd3, mystd4, mystd5]
```
